```python
import keras
import random
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from keras.callbacks import EarlyStopping
from keras.layers import Dropout
from keras.layers import Dense
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint

```

```python
Definecolumn=['VendorID','passenger_count',
              'trip_distance','RatecodeID','store_and_fwd_flag','PULocationID','DOLocationID','payment_type','fare_amount',
              'extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount','congestion_surcharge','Days',
              'tavg','tmin','tmax','wspd','pres']
Targetcolumn=['trip_duration']              
```

```python
X = NewDf[Definecolumn].values
Y = NewDf[Targetcolumn].values
```

```python
DefineScaler=StandardScaler()
TargetScaler=StandardScaler()
```

```python
DefineScalerFit=DefineScaler.fit(X)
TargetScalerFit=TargetScaler.fit(Y)
```

```python
X=DefineScalerFit.transform(X)
Y=TargetScalerFit.transform(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
#model = Sequential()
```

# Linear

# With optimizer Adam and Learning Rate 0.001

```python
model = keras.Sequential()
model.add(keras.Input(shape=(22)))
model.add(Dense(1, activation="relu"))

```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=0.001),loss='mse', metrics='mse')
```

```python

model.summary()
```

```python
history = model.fit(X_train,Y_train, batch_size = 32, epochs=100, validation_data=(X_test,Y_test),verbose=True, shuffle=True)
```

```python
history_NewDf1 = pd.DataFrame(history.history)
plt.plot(history_NewDf1['loss'], label='loss')
plt.plot(history_NewDf1['val_loss'], label='val_loss')

plt.legend()
```

```python
model.evaluate(X_test, Y_test, batch_size=32)
```

# With optimizer SGD 

```python
from tensorflow.keras.optimizers import SGD
modelsgd = keras.Sequential()
modelsgd.add(keras.Input(shape=(22)))
modelsgd.add(Dense(1, activation="relu"))
modelsgd.compile(loss = 'mean_squared_error', optimizer = 'sgd', metrics=['mae'])
history = modelsgd.fit(X_train,Y_train, batch_size = 32, epochs=100, validation_data=(X_test,Y_test),verbose=True, shuffle=True)
```

```python
history_NewDf1 = pd.DataFrame(history.history)
plt.plot(history_NewDf1['loss'], label='loss')
plt.plot(history_NewDf1['val_loss'], label='val_loss')

plt.legend()
```

```python
modelsgd.evaluate(X_test, Y_test, batch_size=32)
```

# MLP

```python
#Multilayered neural network architecture
model1 = Sequential()
model1.add(Dense(15, input_dim=22, activation='relu'))
model1.add(Dense(1, activation='linear'))
model1.summary()
```

```python
model1.compile(loss='mse', optimizer='adam', metrics=['mse', 'mae'])
```

```python
history = model1.fit(X_train, Y_train, epochs=100, validation_split=0.2)
```

```python
history_NewDf = pd.DataFrame(history.history)
plt.plot(history_NewDf['loss'], label='loss')
plt.plot(history_NewDf['val_loss'], label='val_loss')

plt.legend()
```

```python
model1.evaluate(X_test, Y_test, batch_size=32)
```

```python
y_pred = model1.predict(X_test).flatten()

a = plt.axes(aspect='equal')
plt.scatter(Y_test, y_pred)
plt.xlabel('True values')
plt.ylabel('Predicted values')
plt.title('A plot that shows the true and predicted values')
plt.xlim([0, 6])
plt.ylim([0, 6])
plt.plot([0, 6], [0, 6])
```



# Optimizer= RMSProp with Learning rate(lr)=0.0005

```python
import tensorflow as tf

modelprop = Sequential()
modelprop.add(Dense(15, input_dim=22, activation='relu'))
modelprop.add(Dense(1, activation='linear'))
modelprop.summary()
modelprop.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0005),loss='mse',metrics=['mae'])
history = modelprop.fit(X_train, Y_train, epochs=100, validation_data=(X_test, Y_test), batch_size = 32, verbose=1, shuffle=True)
```

```python
history_NewDf = pd.DataFrame(history.history)
plt.plot(history_NewDf['loss'], label='loss')
plt.plot(history_NewDf['val_loss'], label='val_loss')

plt.legend()
```

```python

modelprop.evaluate(X_test, Y_test, batch_size=32)
```
